# Unsupervise Learning

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE(n_components=2,random_state=11)
```

### Transforming the digits Dataset's Features into Two Dimentions

```python
reduced_data=tsne.fit_transform(digits.data)
```

```python
reduced_data.shape
```

### Visualizing the Reduced Data

```python
import matplotlib.pyplot as plt
```

```python
figure=plt.figure(figsize=(5,5))
```

```python
dots=plt.scatter(reduced_data[:,0],reduced_data[:,1],c='black')
```

### Visualizing the Reduced Data with Different Colors for Each Digit

```python
dots=plt.scatter(reduced_data[:,0],reduced_data[:,1],c=digits.target,cmap=plt.cm.get_cmap('nipy_spectral_r',10))
```

```python
colorbar=plt.colorbar(dots)
```

### Vidualizing the reduced data in 3D

```python
tsne3=TSNE(n_components=3,random_state=11)
reduced_data3=tsne3.fit_transform(digits.data)
import mpl_toolkits.mplot3d as axes3d
```

```python
axes=figure.add_subplot(projection='3d')
```

```python
dots=axes.scatter(reduced_data3[:,0],reduced_data3[:,1],reduced_data3[:, 2],c=digits.target,cmap=plt.cm.get_cmap('nipy_spectral_r',10))
```

```python
colorbar=plt.colorbar(dots)
```
